# Q1) Show the number of Confirmed,Deaths and Recovered cases in each region

```python
data.head(2)
```

```python
# data.groupby('Region').sum().head(20)
```

# Q2) Remove all the records where confirmed cases are less than 10

```python
data.head(2)
```

```python
data = data[~(data.Confirmed<10)]   # To remove the records satisfying a particular condition
```

```python
data.head(20)
```

# Q3) In which region, maximum number of Deaths cases were recovered?

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=0).head(30)
```

# Q4) In which region, minimum number of Deaths cases were recovered?

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=1).head(30)
```
